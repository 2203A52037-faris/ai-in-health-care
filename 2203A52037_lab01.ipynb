```python
#i. Y = 3X +2,
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X = np.random.rand(100, 1)
Y = 3 * X + 2 + np.random.normal(0, 0.1, (100, 1))

dataset = tf.data.Dataset.from_tensor_slices((X, Y))
dataset = dataset.shuffle(buffer_size=100).batch(10)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,))
])

model.compile(optimizer='sgd', loss='mean_squared_error')

history = model.fit(dataset, epochs=100, verbose=0)

weights = model.layers[0].get_weights()
print("Weights:", weights[0])
print("Bias:", weights[1])

X_test = np.array([[0], [1]])
Y_pred = model.predict(X_test)

plt.scatter(X, Y, label='Data')
plt.plot(X_test, Y_pred, color='red', label='Model')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
X1 = np.random.rand(100, 1)
X2 = np.random.rand(100, 1)
X = np.hstack((X1, X2))
Y = 3 * X1 + 2 * X2 + 1 + np.random.normal(0, 0.1, (100, 1))

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2,))
])

model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(X, Y, epochs=100, verbose=0)

weights = model.layers[0].get_weights()
print("Weights:", weights[0])
print("Bias:", weights[1])

X_test = np.array([[0.5, 0.5], [1.0, 1.0]])
Y_pred = model.predict(X_test)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X1, Y, label='Data (X1)')
plt.plot(X1, model.predict(X), color='red', label='Model', linestyle='--')
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('X1 vs Y')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(X2, Y, label='Data (X2)', color='orange')
plt.plot(X2, model.predict(X), color='red', label='Model', linestyle='--')
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('X2 vs Y')
plt.legend()

plt.tight_layout()
plt.show()

print("Predictions:")
for (x_test, y_pred) in zip(X_test, Y_pred):
    print(f"Input: {x_test}, Predicted Output: {y_pred[0]}")
```
